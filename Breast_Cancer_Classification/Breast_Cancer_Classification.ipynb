### Step 2 -> Building the layers in the neural network/model

###### Whenever we are training a neural network we need to flatten the data (making it in a single dimensional array from a multi dimensional array like X_train,Y_train,etc)

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),  # here we dont put the target column , only the input features
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

### Step 3 -> Compiling the neural network/model

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-3),loss=keras.losses.SparseCategoricalCrossentropy(),metrics=['accuracy'])
```

### Step 4 -> Training the neural network/model

###### we use history since we will plot this , it's a conventional name

```python
history = model.fit(X_train,y_train,validation_split=0.1,epochs=30)
```

### We have not standardized the data here, we will try by doing it if it is increasing the accuracy

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train) # we are fitting the model to standard scaler on training data and transforming it
# we do not want to fit it in test data otherwise data leakage will happen
X_test_std = scaler.transform(X_test)
```

```python
X_train
```

```python
X_train_std
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
]
)
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(X_train_std,y_train,validation_split=0.1,epochs=10)
```

#### Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend(['training data','validation data'],loc='lower right')
```

### Step 5 -> Accuracy of the model on test data

```python
loss,accuracy = model.evaluate(X_test_std,y_test)
```

### Step 8 -> Building the predictive system for 1 data point(it will tell the output when we give value)

```python
input_data = (13,21.82,87.5,519.8,0.1273,0.1932,0.1859,0.09353,0.235,0.07389,0.3063,1.002,2.406,24.32,0.005731,0.03502,0.03553,0.01226,0.02143,0.003749,15.49,30.73,106.2,739.3,0.1703,0.5401,0.539,0.206,0.4378,0.1072)

# change the input data as numpy array
input_as_np_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for one data point
input_data_reshape = input_as_np_array.reshape(1,-1)

#standardising the input data
input_data_std = scaler.transform(input_data_reshape)

#predicting the values
prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0]==0):
    print("tumour is malignant")
else:
    print("the tumour is benign")
```
